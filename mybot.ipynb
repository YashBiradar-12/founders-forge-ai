```python
def ask_AI(question):
    response = ollama.chat(
        model='mistral',
        messages=[
            {
                'role': 'user',
                'content': question
            }
        ]
    )
    return response['message']['content']
```

```python
answer = ask_AI("What is the tallest mountain in the world?")
print(answer)
```

```python
def gym_trainer(question):
    response = ollama.chat(
        model = 'mistral',
        messages=[
            {
                'role' : 'system',
                'content' : """ You are a gym trainer. 
                rules:
                    - Be polite and friendly
                    - keep answers short and to the point
                    - give diet plan for weigt loss
                """
            },
            {
                'role' : 'user',
                'content' : question
            }
        ]
    )
    return response['message']['content']
```

```python
gym_trainer("I want to lose weight. What should I do?")
```
